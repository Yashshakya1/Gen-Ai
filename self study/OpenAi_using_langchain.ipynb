# Chat Model

* Chat models are language models that use a sequence of messages as inputs and return messages as outputs (as opposed to traditional, plaintext LLMs).

# use of chat model 
* Chatbots
* Customer support AI
* Coding help
* Learning
* Multi-turn reasoning
* Complex instructions

# 🚀 1) Tool Calling
* Tool Calling allows an AI model to use external tools, APIs, functions, or databases to get real results.
* ex
* {
  "tool": "get_weather",
  "city": "Ujjain"
}

# 📦 2) Structured Output
* What it means
* Structured Output means the AI returns answers in a fixed, machine-readable format, like:
* JSON
* dictionaries
* lists
* tables
* HTML
* XML
* It ensures the output never changes shape.

# 🖼️ 3) Multimodality
* What it means
* Multimodality =
* An AI model can understand multiple types of input, not just text.
* This includes:
* Images
* Audio
* Video (in some models)
* PDFs
* Screenshots
* Handwriting


| Feature               | What It Means                  | Why It Matters          |
| --------------------- | ------------------------------ | ----------------------- |
| **Tool Calling**      | AI can use external tools/APIs | Real data, real actions |
| **Structured Output** | AI gives fixed-format output   | Reliable for coding     |
| **Multimodality**     | AI can understand images/audio | Real-world intelligence |
