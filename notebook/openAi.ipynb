# 🔥 Temperature = Creativity Control
| Temperature   | Model Behavior                                               |
| ------------- | ------------------------------------------------------------ |
| **0.0**       | बिल्कुल accurate, fixed, predictable answers (no creativity) |
| **0.2**       | हल्का variation, mostly accurate                             |
| **0.5**       | थोड़ा creative + balanced                                    |
| **0.7 – 1.0** | ज्यादा creative, नए ideas                                    |
| **>1.0**      | बहुत random, unpredictable                                   |


```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "what is dl"}
    ],
    temperature=0.2  # spelling fix
)

print(response.choices[0].message.content)
```
