# ✅ What is RAG?
* RAG stands for Retrieval-Augmented Generation.
* It is a method used in AI to make answers more accurate, reliable, and fact-based.

# ✅ Simple Meaning
* RAG = Search + AI Answering
* The AI first retrieves information from documents or a database,
* then it generates an answer using that information.

# ✅ Meaning of Hallucination in AI
* When an AI gives incorrect, made-up, or false information while sounding confident.

# 📌 Simple Example
 You ask an AI:
* “Who invented the Sun?”
 * AI replies:
* “The Sun was invented by John...”
* (This is wrong and invented → hallucination)

# 🎯 Why RAG Is Important
* Reduces AI hallucinations
* Gives fact-checked answers
* Works with your own data (PDFs, website, database)
* Useful for chatbots, company tools, customer support, etc.
