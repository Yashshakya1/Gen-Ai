# Agent vs RAG

## Comparison Table

| Feature              | RAG      | Agent      |
|----------------------|----------|------------|
| Document Retrieval   | ✅ Yes   | ✅ Yes     |
| Tool Usage           | ❌ No    | ✅ Yes     |
| API Calls            | ❌ No    | ✅ Yes     |
| Reasoning            | Limited  | Multi-step |

---

## RAG (Retrieval-Augmented Generation)
- Knowledge base se relevant documents retrieve karta hai
- Simple Q&A tasks ke liye best
- Static knowledge — sirf documents pe depend karta hai
- Fast & predictable responses

## Agent
- Tools, APIs, calculators use kar sakta hai
- Multi-step reasoning karta hai (ReAct loop)
- Dynamic — real-time data fetch kar sakta hai
- Complex tasks solve karta hai autonomously

# Calculator Tool

````python
from langchain.agents import create_agent
from langchain.tools import tool 
from langchain_ollama import ChatOllama
````

````python
@tool 
def calculator(expr: str) -> str:
    """ evalute a mathematical expression """
    return str(eval(expr))
````

# Load LLm

````python
llm = ChatOllama(model= 'llama3.1:8b')
````

# create agent 

````python
agent = create_agent(
    model= llm,
    tools=[calculator],
    system_prompt= "you are a helpfull assistant"
)
````

# invoke 

````python
response = agent.invoke({"messages":[{"role":"user","content": "25*4+10"}]})
print(response)
````

````python
print(response["messages"][-1].content)
````

# 2 tool Word Count tool

````python
# from langchain.tools import tool
# from langchain.agents import tool
````

# define Tool

````python
@tool
def ward_count(text:str) -> str:
    """ Count the number of words in a given text."""
    return str(len(text.split()))
````

````python
# llm 
llm = ChatOllama(model= 'llama3.1:8b')
````

# create agent 

````python
agent = create_agent(
    model=llm,
    tools=[ward_count],
    system_prompt="Use the ward_count tool when user ask to count words. Return only the number"
)
````

# invoke 

````python
response = agent.invoke({
    "messages": [{"role": "user","content":"count words in: Langchain is very powerfull"}]
})
````

# print only answer 

````python
print(response["messages"][-1].content)
````
