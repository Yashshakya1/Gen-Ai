```python
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

# Create Embedding Model 

```python
embeddings = OllamaEmbeddings(
    model = "nomic-embed-text"
)
```

# Load a Documet 

```python
loader = PyMuPDFLoader("/Users/yashshakya787/genAI/notebook/sample_ai_document.pdf")
docs = loader.load()
len(docs)
```

Split the Document into chunks 

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)
len(chunks)
```

# create Faiss vector store

```python
faiss_db = FAISS.from_documents(
    chunks,embeddings
)
```

# search similar documents with Faiss

```python
query = "what is ai"
result = faiss_db.similarity_search(query)
result[0].page_content
```

# convert Faiss into retriever

```python
retriever = faiss_db.as_retriever()
retriever.invoke("give me a funny jokes for the girls ")
```

Part 2 ChoromaDb Vec
chroma is a presist based

```python
chroma_db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="../data/chroma_db"
)
```

# Search Documents with ChromaDb

```python
query = "explain ai"
results = chroma_db.similarity_search(query)

print(results[0].page_content)
```

# create Retriever from chromadb

```python
retriever = chorma_db.as_retriever()
retriever.invoke("what is ml")
```
