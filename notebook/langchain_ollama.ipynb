```python
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
load_dotenv()
base_url = "http://localhost:11434"
model = 'llama3.1:8b'

llm = ChatOllama(
    base_url=base_url,
    model = model,
    temperature  = 0.8,
    num_predict = 256
)

response = llm.invoke('What is generative ai?')
print(response.content)
```
