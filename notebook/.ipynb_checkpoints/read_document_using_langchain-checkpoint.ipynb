connect 

```python
engine = create_engine("sqlite:////Users/yashshakya787/genAI/notebook/sales.db")
db = SQLDatabase(engine)
```

```python
loader = SQLDatabaseLoader(
    query = "SELECT * FROM sales",
    db = db
)
docs = loader.load()
```

```python
len(docs)
```
