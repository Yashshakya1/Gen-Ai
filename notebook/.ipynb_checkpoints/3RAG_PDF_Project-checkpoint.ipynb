# RAG PDF Question Answering System
This notebook loads a PDF, creates embeddings, stores them in FAISS, and answers questions.

```python
import os
import numpy as np
import faiss
from dotenv import load_dotenv
from openai import OpenAI
from pypdf import PdfReader
```

```python
# Load API Key
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ''
    for page in reader.pages:
        text += page.extract_text() + '\n'
    return text
```

```python
def chunk_text(text, chunk_size=500, overlap=100):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks
```

```python
def get_embeddings(texts):
    response = client.embeddings.create(
        model='text-embedding-3-small',
        input=texts
    )
    return np.array([d.embedding for d in response.data])
        .astype('float32')
```

```python
def create_faiss_index(embeddings):
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    return index
```

```python
def search(query, index, chunks, k=3):
    query_embedding = get_embeddings([query])
    distances, indices = index.search(query_embedding, k)
    return [chunks[i] for i in indices[0]]
```

```python
def ask_question(query, index, chunks):
    relevant_chunks = search(query, index, chunks)
    context = '\n\n'.join(relevant_chunks)

    prompt = f'''Answer the question based only on the context below.

Context:
{context}

Question:
{query}'''

    response = client.chat.completions.create(
        model='gpt-4o-mini',
        messages=[
            {'role': 'system', 'content': 'You are a helpful assistant.'},
            {'role': 'user', 'content': prompt}
        ]
    )
    return response.choices[0].message.content
```

```python
# MAIN EXECUTION
pdf_text = read_pdf('../data/Dr_Ashish_Chandiok_RAG_Resume.pdf')
chunks = chunk_text(pdf_text)
embeddings = get_embeddings(chunks)
index = create_faiss_index(embeddings)

print('RAG Ready!')
```

```python
while True:
    question = input("\nAsk a question (type 'exit' to stop): ")

    if question.lower() == "exit":
        break

    answer = ask_question(question, index, chunks)
    print("\nAnswer:", answer)
```
