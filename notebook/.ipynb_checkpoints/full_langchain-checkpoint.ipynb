```python
import os 
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
load_dotenv()
```

```python
llm = ChatOpenAI(
    model = "gpt-4o-mini",
    temperature=0
)

prompt = ChatPromptTemplate.from_template(
    "Expain {topic} in complex wards"
)

#create chain

chain = prompt | llm 
# invoke chain
response_1 = chain.invoke({"topic : Machine Learning"})
response_2 = chain.invoke({"topic : toxic girls"})
print(response_1.content)
print("2nd prompt\n",response_2.content)
```

```python
import os 
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
load_dotenv()
```

```python
llm = ChatOpenAI(model = "gpt-4o_mini", temperature=0)

parser = StrOutputParser()

explain_prompt = ChatPromptTemplate.from_template(
    "explain {topic} in details"
)
summary_prompt = ChatPromptTemplate.from_template(
    "summarize the following text in 3 lines :\n\n{input}"
)
bullet_prompt = ChatPromptTemplate.from_template(
    "convert the following summary into bullet points: \n\n{input}"
)
# create chain 
chain = (
    explain_prompt
    | llm
    | parser
    | summary_prompt
    | llm
    | parser
    | bullet_prompt 
    | llm 
    | parser
)
response = chain.invoke({"topic": "Machine Learning"})
print(response)
```

```python
import os 
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini",temperature = 0)

# output parser (converts AImessage -> string)
parser = StrOutputParser()

explain_prompt = ChatPromptTemplate.from_template(
    "Expalin {topic} in details"
)

summary_prompt = ChatPromptTemplate.from_template(
    "Summarize the following text in 3 line:\n\n{input}"
)

bullet_prompt = ChatPromptTemplate.from_template(
    "Convert the following summary into bullet points:\n\n{input}"
)

# proper 3 step chain
chain = (
    explain_prompt 
    | llm 
    | parser
    | summary_prompt 
    | llm 
    | parser
    | bullet_prompt  
    | llm 
    | parser
)

# run 
response = chain.invoke({"topic": "toxic girls"})

print(response)
```
