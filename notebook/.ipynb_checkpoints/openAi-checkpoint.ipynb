# mcp 
* MCP = AI ko real-world tools se connect karne ka standard method.
* 
<!-- gen ai wrokflow -->


| Feature                 | Explanation                                          |
| ----------------------- | ---------------------------------------------------- |
| **Tools se connection** | AI directly APIs, databases, files use kar sakta hai |
| **Secure**              | Controlled access hota hai                           |
| **Standard protocol**   | Sab AI platforms ek hi system follow karte hain      |
| **Reusable**            | Ek baar tool bana do → koi bhi AI use kar sakta hai  |

* RAG (Retrieval Augmented Generation)
* Agent systems
* Workflow automation
* AI coding tools
* Enterprise AI integration

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "what is dl"}
    ],
    temperature=0.2  # spelling fix
)

print(response.choices[0].message.content)
```
