# The Simplest Graph

Let's build a simple graph with 3 nodes and one conditional edge.
```
Start → Node 1 → Node 2 → End
               ↘ Node 3 ↗
```

- **Node 1** — entry point, decides which path to take
- **Node 2** — path A (conditional)
- **Node 3** — path B (conditional)
- Both Node 2 & Node 3 lead to **End**

## State

First, define the State of the graph.

The State schema serves as the input schema for all Nodes and Edges in the graph.

Let's use the `TypedDict` class from python's `typing` module as our schema, 
which provides type hints for the keys.

````python
from typing_extention import Typeict
class State(typeDict):
    graph_state: str
````

````python
def node_1(state):
    print("----Node 1----")
    return {"graph_state":state["graph_state"] + "I am"}

def node_2(state):
    print("----Node 2----")
    return {"graph_state":state["graph_state"] + "I am "}

def node_3(state):
    print("----Node 3----")
    return {"graph_state":state["graph_state"] + "Sad"}
````
