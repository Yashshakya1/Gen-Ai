```python
from langchain_ollama import OllamaEmbeddings,ChatOllama
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
```

```python
embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
)
```

# embed a single text

```python
text = "ai is transfroming the world"
vector = embedding.embed_query(text)

print("vector length:", len(vector))
print(vector[:10])
```

```python
documents = [
    "Artificial Intelligence is transforming industries",
    "Machin learning enables system to learn from data ",
    "Deep learning uses neural networks"
]
vector = embeddings.embed_documents(documents)
print(len(vector))
```

# Load a pdf document 

```python
loader = PyMuPDFLoader("/Users/yashshakya787/genAI/notebook/sample_ai_document.pdf")
docs = loader.load()
print(len(docs))
```

# split Documents into Chunks 

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)
chunks = splitter.split_documents(docs)
print(len(chunks))
```

# Create Embedding for Chunks

```python
vectors = embeddings.embed_documents(
    [chunks.page_content for chunks in chunks]
)
print(len(vectors))
```

Store Embeddings in Faiss store the data 

```python
vectorstore = FAISS.from_documents(
    chunks,
    embedding
)
```

Search similar Documents

```python
# query = "what is AI"
# result = vectorstore.similarity_search(query)
# print(results[0]).page_content()


query = "what is AI"
results = vectorstore.similarity_search(query)

print(results[0].page_content)
```
